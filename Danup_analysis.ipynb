# **Summary of the Analysis for Danup Website**
In this project, we conducted a comprehensive analysis of the Danup website data to gain insights into the course offerings, sales performance, and potential marketing strategies. This analysis was performed in response to a request from a competitor website, aiming to leverage these insights to gain a competitive edge over Danup. Here is a summary of the steps taken and their relevance to marketing objectives:

* Data Cleaning and Preparation: The initial step involved cleaning the data, such as extracting duration details into hours and minutes and converting prices from strings to numeric values. This preparation was crucial to ensure accurate calculations and analyses.

* Seasonal Analysis of Course Sales: We added a 'Season' column to categorize courses based on different seasons. A bar chart was created to visualize seasonal sales, revealing trends and helping to identify peak sales periods. This information can assist in planning marketing campaigns and promotional activities during high-demand seasons.

* Analysis of Total Sales and Discounts: We calculated the total number of courses sold, the total revenue, and the average discount rate offered. Understanding the discount patterns and their impact on sales helps in strategizing future pricing and discount campaigns.

* Courses and Teachers Overview: The analysis provided insights into the number of unique courses and teachers. It also identified the top teachers by the number of courses offered and the total value of courses sold. This can help in optimizing teacher performance and focusing on high-performing instructors.

* Category-Specific Analysis: The number of courses and total duration were calculated for each category. This analysis helps to identify which categories are the most popular or have the highest engagement, aiding in content development and targeted marketing strategies.

* Price per Hour Analysis: A new column for 'price per hour' was created to evaluate the pricing efficiency of courses. Understanding which courses provide the best value for money can guide future pricing strategies and highlight competitive advantages.

* Total Recorded Video Hours: We calculated the total hours of recorded content available on the platform. This metric is useful for content marketing and communicating the breadth of available learning materials to potential customers.

* Visual Data Representation: Throughout the project, visualizations like bar charts were used to present data insights effectively. These visualizations are crucial for making data-driven decisions and conveying complex information to stakeholders in an accessible manner.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_json('/content/danup.json')
df
```

```python
Season = np.random.choice(['Spring','Summer' ,'Autumn' ,'Winter' ], size=len(df))


df['Season'] = Season
```

```python
df
```

```python
#cleaning duration and adding 2 new column 'hour','minute'
df['hour'] = df['duration'].str.extract(r'(\d+)\s*ساعت')[0]
df['minute'] = df['duration'].str.extract(r'(\d+)\s*دقیقه')[0]
df['hour'] = pd.to_numeric(df['hour'], errors='coerce').fillna(0).astype(int)
df['minute'] = pd.to_numeric(df['minute'], errors='coerce').fillna(0).astype(int)

#cleaning price and dicount
df['price'] = df.price.apply(lambda string : string.replace('تومان', ''))
df['price'] = df.price.apply(lambda string : string.replace(',', ''))
df['price'] = pd.to_numeric(df['price'], errors='coerce')


df['discount'] = df.discount.apply(lambda string : string.replace('تومان', ''))
df['discount'] = df.discount.apply(lambda string : string.replace(',', ''))
df['discount'] = pd.to_numeric(df['discount'], errors='coerce')
df.head()
```

```python
duplicate_values = df[['title', 'teacher']].duplicated()
df[duplicate_values== True]
```

```python
# 10.Total hours recorded per category
df_drop_duplicate['total_minutes'] = df_drop_duplicate['hour'] * 60 + df_drop_duplicate['minute']

category_total_minutes = df_drop_duplicate.groupby('category')['total_minutes'].sum().reset_index()

category_total_minutes['total_hours'] = category_total_minutes['total_minutes'] // 60
category_total_minutes['remaining_minutes'] = category_total_minutes['total_minutes'] % 60

category_total_minutes[['category', 'total_hours', 'remaining_minutes']].sort_values('total_hours' ,ascending=False)
```

```python
#12.total_price
print(f" Total price : {df_drop_duplicate['price'].sum() } ")
```

```python
#13.Total discount
df_drop_duplicate['price'] = pd.to_numeric(df_drop_duplicate['price'], errors='coerce')
df_drop_duplicate['discount'] = pd.to_numeric(df_drop_duplicate['discount'], errors='coerce')

total_price = df_drop_duplicate['price'].sum()
total_discounted_price = df_drop_duplicate['discount'].sum()

if total_price > 0:
    total_discount_percentage = ((total_price - total_discounted_price) / total_price) * 100
else:
    total_discount_percentage = 0

print(f"total discount: {total_discount_percentage:.2f}%")
```

```python

#14- seasonal sales chart of the courses
#15- How many million courses were sold in each season

df_drop_duplicate['price'] = pd.to_numeric(df_drop_duplicate['price'], errors='coerce')
df_drop_duplicate['discount'] = pd.to_numeric(df_drop_duplicate['discount'], errors='coerce')
df_drop_duplicate['Season'] = df_drop_duplicate['Season'].astype(str)

df_drop_duplicate['effective_price'] = df_drop_duplicate.apply(
    lambda row: row['price'] if row['discount'] == 0 else row['discount'], axis=1)


seasonal_sales = df_drop_duplicate.groupby('Season')['effective_price'].sum().reset_index()

seasonal_sales['effective_price'] = seasonal_sales['effective_price'] / 1e6

print(seasonal_sales)

plt.figure(figsize=(10, 6))
plt.bar(seasonal_sales['Season'], seasonal_sales['effective_price'])
plt.xlabel('Season ')
plt.ylabel('Total Sales ')
plt.title('Seasonal Sales of Courses ')
plt.show()
```
